```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Excel file
file_path = 'Data/Population_Size.xlsx'
df_population_size = pd.read_excel(file_path, sheet_name='Sheet1')
```

```python
# Cleaning up the data
df_population_size['Population Size'] = df_population_size['Population Size'].fillna(method='ffill')
df_population_size_cleaned = df_population_size[['Population Size', 'Production Efficiency']]

# Grouped by population size and calculated minimum, maximum and average values of productivity
grouped = df_population_size_cleaned.groupby('Population Size').agg(['min', 'max', 'mean']).reset_index()

# Flatten column names
grouped.columns = ['Population Size', 'Min Efficiency', 'Max Efficiency', 'Mean Efficiency']

# Force conversion of data types to floating point numbers
grouped = grouped.astype(float)

# Plotting line graphs
plt.figure(figsize=(10, 6))

# Plotting the average efficiency line
plt.plot(grouped['Population Size'], grouped['Mean Efficiency'], label='Mean Efficiency', color='skyblue', marker='o')

# Plot error bars for min and max values, enhance visibility of grey bars
plt.errorbar(grouped['Population Size'], grouped['Mean Efficiency'], 
             yerr=[grouped['Mean Efficiency'] - grouped['Min Efficiency'], grouped['Max Efficiency'] - grouped['Mean Efficiency']], 
             fmt='o', color='black', ecolor='darkgray', elinewidth=5, capsize=5, label='Efficiency Range (Min to Max)')

# Add tags and titles
plt.xlabel('Population Size')
plt.ylabel('Production Efficiency')
plt.title('Production Efficiency by Population Size')
plt.legend()

# Display grids and charts
plt.grid(True)
plt.show()
```

```python
# Cleaning up the data
df_population_size['Population Size'] = df_population_size['Population Size'].fillna(method='ffill')
df_cleaned = df_population_size[['Population Size', 'Time Period']]

# Grouped by population size and calculated minimum, maximum and average values of productivity
grouped_time = df_cleaned.groupby('Population Size').agg(['min', 'max', 'mean']).reset_index()

# Flatten column names
grouped_time.columns = ['Population Size', 'Min Time', 'Max Time', 'Mean Time']

# Force conversion of data types to floating point numbers
grouped_time = grouped_time.astype(float)

# Plotting line graphs
plt.figure(figsize=(10, 6))

# Plotting the average efficiency line
plt.plot(grouped_time['Population Size'], grouped_time['Mean Time'], label='Mean Time', color='skyblue', marker='o')

# Plot error bars for min and max values, enhance visibility of grey bars
plt.errorbar(grouped_time['Population Size'], grouped_time['Mean Time'], 
             yerr=[grouped_time['Mean Time'] - grouped_time['Min Time'], grouped_time['Max Time'] - grouped_time['Mean Time']], 
             fmt='o', color='black', ecolor='darkgray', elinewidth=5, capsize=5, label='Time Range (Min to Max)')

# Add tags and titles
plt.xlabel('Population Size')
plt.ylabel('Time Period')
plt.title('Running Time by Population Size')
plt.legend()

# Display grids and charts
plt.grid(True)
plt.show()
```

```python
# Load the Excel file
file_path = 'Data/Population_Size.xlsx'
df_updated = pd.read_excel(file_path, sheet_name='Sheet1')

# Cleaning up the data
df_updated['Population Size'] = df_updated['Population Size'].fillna(method='ffill')
df_cleaned = df_updated[['Population Size', 'Production Efficiency']]
df_time_cleaned = df_updated[['Population Size', 'Time Period']]

# Grouped by population size and calculated minimum, maximum and average values of productivity
grouped = df_cleaned.groupby('Population Size').agg(['min', 'max', 'mean']).reset_index()
grouped.columns = ['Population Size', 'Min Efficiency', 'Max Efficiency', 'Mean Efficiency']

grouped_time = df_time_cleaned.groupby('Population Size').agg(['min', 'max', 'mean']).reset_index()
grouped_time.columns = ['Population Size', 'Min Time', 'Max Time', 'Mean Time']

# Plot two-coordinate graphs of Production Efficiency and Running Time
fig, ax1 = plt.subplots(figsize=(12, 8))

# Mapping Production Efficiency
color = 'skyblue'
ax1.set_xlabel('Population Size')
ax1.set_ylabel('Production Efficiency', color=color)
ax1.plot(grouped['Population Size'], grouped['Mean Efficiency'], color=color, marker='o', label='Mean Efficiency')
ax1.errorbar(grouped['Population Size'], grouped['Mean Efficiency'], 
             yerr=[grouped['Mean Efficiency'] - grouped['Min Efficiency'], grouped['Max Efficiency'] - grouped['Mean Efficiency']], 
             fmt='o', color='black', ecolor='darkgray', elinewidth=5, capsize=5, label='Efficiency Range (Min to Max)')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='upper left')

# Create a second y-axis for running time
ax2 = ax1.twinx()
color = 'orange'
ax2.set_ylabel('Running Time', color=color)
ax2.plot(grouped_time['Population Size'], grouped_time['Mean Time'], color=color, marker='o', label='Mean Time')
ax2.errorbar(grouped_time['Population Size'], grouped_time['Mean Time'], 
             yerr=[grouped_time['Mean Time'] - grouped_time['Min Time'], grouped_time['Max Time'] - grouped_time['Mean Time']], 
             fmt='o', color='black', ecolor='darkgray', elinewidth=5, capsize=5, label='Time Range (Min to Max)')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper right')

# Add overall chart title
plt.title('Production Efficiency and Running Time by Population Size')

# Display grids and charts
# Adjust subgraphs to fit the graphic area
fig.tight_layout()
plt.grid(True)
plt.show()
```

```python
file_path = 'Data/Population_Size.xlsx'
df_updated = pd.read_excel(file_path, sheet_name='Sheet1')

df_updated['Population Size'] = df_updated['Population Size'].fillna(method='ffill')
df_cleaned = df_updated[['Population Size', 'Production Efficiency']]
df_time_cleaned = df_updated[['Population Size', 'Time Period']]

# Grouped by population size and calculated minimum, maximum and average values of productivity
grouped = df_cleaned.groupby('Population Size').agg(['min', 'max', 'mean']).reset_index()
grouped.columns = ['Population Size', 'Min Efficiency', 'Max Efficiency', 'Mean Efficiency']

grouped_time = df_time_cleaned.groupby('Population Size').agg(['min', 'max', 'mean']).reset_index()
grouped_time.columns = ['Population Size', 'Min Time', 'Max Time', 'Mean Time']

# Plot two-coordinate graphs of Production Efficiency and Running Time
fig, ax1 = plt.subplots(figsize=(12, 8))

# Mapping Production Efficiency
color = 'skyblue'
ax1.set_xlabel('Population Size')
ax1.set_ylabel('Production Efficiency', color=color)
ax1.plot(grouped['Population Size'], grouped['Mean Efficiency'], color=color, marker='o', label='Mean Efficiency')
ax1.fill_between(grouped['Population Size'], grouped['Min Efficiency'], grouped['Max Efficiency'], color=color, alpha=0.2, label='Efficiency Range (Min to Max)')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='upper left', bbox_to_anchor=(0.5, -0.1))

# Create a second y-axis for running time
ax2 = ax1.twinx()
color = 'orange'
ax2.set_ylabel('Computational Time(s)', color=color)
ax2.plot(grouped_time['Population Size'], grouped_time['Mean Time'], color=color, marker='o', label='Mean Time')
ax2.fill_between(grouped_time['Population Size'], grouped_time['Min Time'], grouped_time['Max Time'], color=color, alpha=0.2, label='Time Range (Min to Max)')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper right', bbox_to_anchor=(0.5, -0.1))

# Add overall chart title
plt.title('Production Efficiency and Computational Time(s) by Different Population Size')

# Display grids and charts
# Adjust subgraphs to fit the graphic area
fig.tight_layout() 
plt.grid(True)
plt.show()
```
