```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_Jan = pd.read_excel('Data/Wind_Directions_Vis_Jan.xlsx')
df_Jul = pd.read_excel('Data/Wind_Directions_Vis_Jul.xlsx')
```

```python
# Map directions to angles in radians, starting with N at the top
direction_angles = {
    'N': np.pi / 2,
    'NE': 3 * np.pi / 4,
    'E': np.pi,
    'SE': 5 * np.pi / 4,
    'S': 3 * np.pi / 2,
    'SW': 7 * np.pi / 4,
    'W': 0,
    'NW': np.pi / 4
}

# Convert directions to angles
angles = df_Jan['Direction'].map(direction_angles).values
values = df_Jan['Probability'].values

# Define the colors from the image
colors = [
    (8/255, 29/255, 89/255),
    (33/255, 49/255, 140/255),
    (34/255, 84/255, 163/255),
    (30/255, 128/255, 184/255),
    (48/255, 165/255, 194/255),
    (91/255, 191/255, 192/255),
    (149/255, 213/255, 184/255),
    (205/255, 235/255, 179/255),
    (254/255, 248/255, 180/255),
    (254/255, 254/255, 215/255)
]

# Create the polar plot
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
bars = ax.bar(angles, values, width=np.pi/4)

# Set the colors and transparency for each bar
for r, bar, color in zip(values, bars, colors):
    bar.set_facecolor(color)
    bar.set_alpha(0.7)

# Set the direction labels
ax.set_xticks(list(direction_angles.values()))
ax.set_xticklabels(list(direction_angles.keys()))

# Add a simple legend outside the plot
ax.legend(['Probability'], loc='upper right', bbox_to_anchor=(1.1, 1.1))

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
df_Jan = pd.read_excel('Data/Wind_Directions_Vis_Jan.xlsx')

# Put the directions in order
ordered_directions = ['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW']
df_Jan['Direction'] = pd.Categorical(df_Jan['Direction'], categories=ordered_directions, ordered=True)
df_Jan = df_Jan.sort_values('Direction')

# Drawing bar graphs
plt.figure(figsize=(10, 6))
bars = plt.bar(df_Jan['Direction'], df_Jan['Probability'], color='skyblue', edgecolor='black')

# Add chart titles and labels
# Increase spacing between titles and charts
plt.title('Wind Direction Probabilities (January)', fontsize=16, pad=20)
plt.xlabel('Wind Direction', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Add probability labels to the top of the bar
for bar in bars:
    yval = bar.get_height()
    # Move the labels down slightly
    # Remove background colour
    plt.text(bar.get_x() + bar.get_width()/2, yval - 0.02, f'{yval:.2f}', 
             ha='center', fontsize=12, color='black', weight='bold')  

plt.show()
```
