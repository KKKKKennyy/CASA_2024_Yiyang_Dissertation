```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Excel file
file_path = 'Data/Turbine_Number_Analysis.xlsx'
df = pd.read_excel(file_path)
```

```python
# Assuming the columns are named 'Turbine Number', 'Power Production Per Turbine', and 'Production Efficiency'
turbine_number = df['Turbine Number']
power_production = df['Power Production Per Turbine (KW)']
production_efficiency = df['Production Efficiency']
Turbine_density=df['Turbine Density (turbine/km2)']

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Power Production Per Turbine
ax1.plot(turbine_number, power_production, color='orange', marker='o', label='Power Production Per Turbine')
ax1.set_xlabel('Turbine Number')
ax1.set_ylabel('Power Production Per Turbine (KW/Year)', color='orange')
ax1.tick_params(axis='y', labelcolor='orange')


# Create a second y-axis for Production Efficiency
ax2 = ax1.twinx()
ax2.plot(turbine_number, production_efficiency, color='skyblue', marker='o', label='Production Efficiency')
ax2.set_ylabel('Production efficiency', color='skyblue')
ax2.tick_params(axis='y', labelcolor='skyblue')


# Add title and legend
plt.title('Power Production Per Turbine (KW/Year) and Production Efficiency by Different Turbine Number')
fig.tight_layout()

# Show the plot
plt.show()
```
